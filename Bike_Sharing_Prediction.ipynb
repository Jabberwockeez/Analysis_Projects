## 数据清洗

1）数据预处理

此数据集不需要

2）特征工程

变量说明：

datatime    日期+时间  
season      1=春 2=夏 3=秋 4=冬  
holiday     1=节假日 0=非节假日  
workingday   1=工作日 0=周末  
weather     1=晴天多云 2=雾天阴天 3=小雪小雨 4=大雨大雪大雾  
temp       气温摄氏度  
atemp      体感温度  
humidity    湿度  
windspeed    风速  
casual      非注册用户个数  
registered   注册用户个数  
count      给定日期时间（每小时）总租车人数  

```python
full['year'] = full['datetime'].apply(lambda x: x.split()[0].split('-')[0]).astype(int)
full['month'] = full['datetime'].apply(lambda x: x.split()[0].split('-')[1]).astype(int)
full['day'] = full['datetime'].apply(lambda x: x.split()[0].split('-')[2]).astype(int)
full['hour'] = full['datetime'].apply(lambda x: x.split()[1].split(':')[0]).astype(int)
```

```python
# periodDf = pd.DataFrame()
# periodDf['date'] = full['datetime'].apply(lambda x: x.split()[0])
# # periodDf['year'] = full['datetime'].apply(lambda x: x.split()[0].split('-')[0]).astype(int)
# periodDf['month'] = full['datetime'].apply(lambda x: x.split()[0].split('-')[1]).astype(int)
# periodDf['day'] = full['datetime'].apply(lambda x: x.split()[0].split('-')[2]).astype(int)
# periodDf['hour'] = full['datetime'].apply(lambda x: x.split()[1].split(':')[0]).astype(int)
# periodDf.drop('date', axis=1, inplace=True)
# # periodDf['weekday'] = full['datetime'].apply(lambda x: x.split(' ')[1].split(':')[0]).astype(int)
# periodDf.head()
```

```python
full.drop(['casual', 'registered'], axis=1, inplace=True)
corrDf = full.corr()
corrDf['count'].sort_values(ascending=False)
```

```python
#选取相关系数>0.15的特征
full_X = full.drop(['datetime', 'day', 'count'], axis=1)
full_X.head()
```

## 构建模型 评估模型

```python
print(full.shape)
full.head()
```

```python
#原始数据集有10886行
sourceRow = 10886
#原始数据集：特征
source_X = full_X.loc[:sourceRow-1,:]
#原始数据集：标签
source_y = full.loc[:sourceRow-1,'count'].astype(int)
#预测数据集：特征
pred_X = full_X.loc[sourceRow:,:]
```

### 使用线性回归算法预测

```python
#建立模型用的训练数据集和测试数据集
from sklearn.model_selection import train_test_split 

train_X, test_X, train_y, test_y = train_test_split(source_X, source_y, train_size=0.8, test_size=0.2)

#导入线性回归算法
from sklearn.linear_model import LinearRegression
#创建模型：线性回归（linear regression）
model = LinearRegression()
#训练模型
model = model.fit(train_X , train_y)
#评估模型
result = model.score(test_X, test_y)
result
```

效果非常一般，下面选择使用随机森林算法预测。

### 使用随机森林算法预测

```python
#导入随机森林算法
# from sklearn import cross_validation
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100).fit(train_X, train_y)
result = model.score(test_X, test_y)
result
```

准确率较高，选择作为最终预测方法。

## 方案实施

```python
#使用机器学习模型，对预测数据集中的数量进行预测
pred_Y = model.predict(pred_X).astype(int)
#生成的预测值是浮点数,转换为整型
# pred_Y = pred_Y.astype(int)
dt = full.loc[sourceRow:,'datetime']
#数据框：日期时间，租车数量
predDf = pd.DataFrame({'datetime': dt,
                       'count': pred_Y})
print(predDf.shape)
predDf.head()
```

```python
#输出到csv文件，上传Kaggle。
predDf = predDf[['datetime', 'count']]
predDf.to_csv('bikesharing_pred.csv', index = False)
```
